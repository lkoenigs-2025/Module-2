```python
#create a tic tac toe board:

player_wins = 0
player2_wins = 0
tie = 0

def main():
  again = 'y'
  while again.lower() == 'y':
    board = []
    row,col = 3, 3
    for r in range(row):
        rows = []
        for c in range(col):
            rows.append(0)
        board.append(rows)

    player(board)
    print('\n               Scoreboard:')
    print('--------------------------')
    print('\nPlayer 1 wins: ',player_wins)
    print('PLayer 2 wins: ',player2_wins)
    print('Number of ties: ',tie)
    again = input('\nWould you like to play again? (y for yes) ')

def print_board(board):
    for r in range(len(board)):
        for c in range(len(board[r])):
            print(board[r][c], end = ' ')
        print()

def player(board):
  global player_wins, player2_wins, tie
  taken = 0
  game = 0
  print('Welcome to the Tic Tac Toe Challenge!')
  while taken != 9 and game != 1:
    for i in ['one','two']:
        print('\nPlayer',i,': \nPlease enter a number from 0 to 2 for a row and column: ',end = '')
        rws_cols = input('(Hint: row,column) ')
        rw = rws_cols.split(",")
        rws = rw[0]
        cls = rw[1]
        tpl = (int(rws), int(cls))
        print(tpl)


        while (tpl[0] > 2 or tpl[0] < 0 or tpl[1] > 2 or tpl[1] < 0) or board[tpl[0]][tpl[1]] == 1 or board[tpl[0]][tpl[1]] == 2:
          if tpl[0] > 2 or tpl[0] < 0 or tpl[1] > 2 or tpl[1] < 0:
            print('\nInput not valid.')
          else:
            print('This spot is already taken.')
          rws_cols = input('Enter a number from 0 to 2 for a row and column: (Hint: row,column)')
          rw = rws_cols.split(",")
          rws = rw[0]
          cls = rw[1]
          tpl = (int(rws), int(cls))

        if i == 'one':
          board[tpl[0]][tpl[1]] = 1
        else:
          board[tpl[0]][tpl[1]] = 2

        taken +=1
        game, player_wins, player2_wins, tie = win_game(board, taken)
        if game == 1:
          break

def win_game(board, taken):
  global player_wins, player2_wins, tie
  row1 = 0
  row2 = 0
  col1 = 0
  col2 = 0
  game = 0

  for i in range(len(board)):
    for j in range(len(board[i])):
      if board[i][j] == 1:
        row1+=1
      if board[i][j] == 2:
        row2+=1
      if board[j][i] == 1:
        col1+=1
      if board[j][i] == 2:
        col2+=1

    if row1 == 3:
      print('\nCongrats player 1- you got a row horizantally!')
      print_board(board)
      player_wins += 1
      game = 1

    if row2 == 3:
      print('\nCongrats player 2- you got a row horizantally!')
      print_board(board)
      player2_wins += 1
      game = 1

    if col1 == 3:
      print('\nCongrats player 1- you got a row vertically!')
      print_board(board)
      player_wins += 1
      game = 1

    if col2 == 3:
      print('\nCongrats player 2- you got a row vertically!')
      print_board(board)
      player2_wins += 1
      game = 1

    row1 = 0
    row2 = 0
    col1 = 0
    col2 = 0

  if board[0][0] == 1 and board[1][1] == 1 and board[2][2] == 1:
    print('\nCongrats player 1- you got a row diagonally!')
    print_board(board)
    player_wins += 1
    game = 1

  if board[0][0] == 2 and board[1][1] == 2 and board[2][2] == 2:
    print('\nCongrats player 2- you got a row diagonally!')
    print_board(board)
    player2_wins += 1
    game = 1

  if board[0][2] == 1 and board[1][1] == 1 and board[2][0] == 1:
    print('\nCongrats player 1- you got a row diagonally!')
    print_board(board)
    player_wins += 1
    game = 1

  if board[0][2] == 2 and board[1][1] == 2 and board[2][0] == 2:
    print('\nCongrats player 2- you got a row diagonally!')
    print_board(board)
    player2_wins += 1
    game = 1

  if taken == 9:
    print('\nOh well, the board is full. You did not get a row.')
    print_board(board)
    tie = 1
    game = 1
  return game, player_wins, player2_wins, tie

main()
```
