## Exercise 2 — `min_max(a, b, c)`
Return `(smallest, largest)` for three integers.


````python
def min_max(a: int, b: int, c: int) -> tuple[int, int]:
    # TODO: compute smallest and largest, return as a tuple
    tpl = (a, b, c)
    smallest = tpl[0]
    largest = tpl[0]
    for i in range(len(tpl)):
      if tpl[i] > largest:
          largest = tpl[i]
      else:
        if tpl[i] < smallest:
          smallest = tpl[i]
    return (smallest, largest)



# Tests
assert_equal(min_max(3, 9, 5), (3, 9), "min_max failed")
assert_equal(min_max(-7, -2, -9), (-9, -2), "min_max failed")
assert_equal(min_max(4, 4, 4), (4, 4), "min_max failed")
print("✅ Exercise 2 passed")
````

## Exercise 7 — `sum_of_quotients_and_remainders(pairs)`
`pairs` is a list of `(n, d)` tuples. Return `(sum_q, sum_r)`.

Example: `[(17,5), (20,4)]` → q's: 3+5=8, r's: 2+0=2.


````python
def sum_of_quotients_and_remainders(pairs: list[tuple[int, int]]) -> tuple[int, int]:
    # TODO:
    # loop through pairs
    # unpack (n, d)
    # get (q, r) from quotient_remainder
    # accumulate and return (sum_q, sum_r)
    sum_q = 0
    sum_r = 0
    for i in range(len(pairs)):
      for j in range(len(pairs[i])):
        n = pairs[i][0]
        d = pairs[i][1]
        q = n // d
        r = n % d
      sum_q += q
      sum_r += r
    return (sum_q, sum_r)


# Tests
assert_equal(sum_of_quotients_and_remainders([(17, 5), (20, 4)]), (8, 2), "sum_of_quotients_and_remainders failed")
assert_equal(sum_of_quotients_and_remainders([(10, 3), (10, 3), (10, 3)]), (9, 3), "sum_of_quotients_and_remainders failed")
print("✅ Exercise 7 passed")
````

## Exercise 8 — `apply_linear(points, m, b)`
Given a list of `(x, y)` points, return a **new** list of transformed points `(x, m*x + b)`.

This is numeric-only, and it's a great tuple unpacking exercise.

**Note:** You should ignore the incoming `y` value; keep `x` and compute a new `y`.


````python
def apply_linear(points: list[tuple[int, int]], m: int, b0: int) -> list[tuple[int, int]]:
    # TODO:
    # For each (x, y) in points, create (x, m*x + b0)
    # Return the new list
    new_list = []
    for i in range(len(points)):
      for j in range(len(points[i])):
        x = points[i][0]
        y = m*x +b0
      tpl = (x, y)
      new_list.append(tpl)
    return new_list

# Tests
pts = [(0, 99), (2, 1), (-3, 7)]
assert_equal(apply_linear(pts, 3, 1), [(0, 1), (2, 7), (-3, -8)], "apply_linear failed")
assert_equal(pts, [(0, 99), (2, 1), (-3, 7)], "apply_linear should not modify original list")
print("✅ Exercise 8 passed")
````

## Exercise 9 — `best_two(scores)`
`scores` is a list of numbers. Return a tuple `(highest, second_highest)`.

Assume the list has **at least 2** elements.

**Hint:** You can track two variables and update them as you scan.

**Requirement:** Return a tuple; do not sort the whole list.


````python
def best_two(scores: list[int]) -> tuple[int, int]:
    # TODO:
    # scan once, track best and second best
    highest = 0
    second_highest = 0
    for i in range(len(scores)):
      for j in range(scores[i]):
        if scores[i][j] > highest:
          highest = scores[i][j]
        else:
          if scores[i][j] > second_highest:
            second_highest = scores[i][j]
    return (highest, second_highest)

# Tests
assert_equal(best_two([5, 1, 9, 3, 9, 2]), (9, 9), "best_two failed")
assert_equal(best_two([-5, -1, -9, -3]), (-1, -3), "best_two failed")
print("✅ Exercise 9 passed")
````
