```python
#WHAT WE CAN DO WITH A TUPLE (and thats it:))
#1. count
tpl = (1,2,3,4)
print(t1.count("l"))
#2. index
print(t1.index("l"))

#3. sum - can only use with int
print("SUM:", sum(tpl))
#4. min - with alphabet it will give u min a=1....
print("MIN:", min(t1))
#5. max - giveu max number or alphabet
print("MAX:", max(t1))
#6. len - size of tuple
print(len(tpl))
```

```python
#is vs equals to
a = (1,2, ["a","b"])
b = (1,2, ["a","b"])
print(a==b)
print(a is b)

a = b
print(f'{a is b}')
a[2].append("c")
print(a) #wld give u same thing cuz u added it to both etc.
print(b)
```

```python
#tuple comprehension
my_tpl = tuple("abcdefg")
my_tuple = tuple( value for value in my_tpl)
print(my_tuple)

my_list = [value for value in my_tpl]
print(my_list)
```

```python
#LOOPS WITH TUPLES

#simple: look at each value within tuple
print(my_tuple)

#every value loop thru
for value in my_tuple: #or can do for i...
  print(value, end='')

#access according by index
print()
for i in range(len(my_tuple)): #length of tuple
  print(my_tuple[i], end ='')

#enumerate
print()
print("Enumerate:")
for i, value in enumerate(my_tuple):
  print(f'{i} {value}')
```

```python
t = ((1,2),(3,4))
print(t)

for i,value in enumerate(t):
  print(i, value)

#or same thing with enumerate:
for i, (val1, val2) in enumerate(t):
  print(i, (val1, val2)) #have to unpack each value as a tuple
```
