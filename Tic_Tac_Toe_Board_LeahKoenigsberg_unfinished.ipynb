```python
#create a tic tac toe board:
board = []
row,col = 3, 3
for r in range(row):
    rows = []
    for c in range(col):
        rows.append(0)
    board.append(rows)

def print_board(board):
    for r in range(len(board)):
        for c in range(len(board[r])):
            print(board[r][c])
        #print(board[r]) - not sure which way to print...
print_board(board)
```
